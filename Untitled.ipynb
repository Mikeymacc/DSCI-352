```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the two Gaussian distributions
mean = 0
variance1 = 1  # Variance of first class
variance2 = 4  # Variance of second class

# Generate data points focusing on the intersection region
x_zoom = np.linspace(-5, 5, 400)
pdf_class1_zoom = norm.pdf(x_zoom, mean, np.sqrt(variance1))
pdf_class2_zoom = norm.pdf(x_zoom, mean, np.sqrt(variance2))
discriminant_zoom = np.log(pdf_class1_zoom / pdf_class2_zoom)

# Plotting the zoomed-in distributions and discriminant function
plt.figure(figsize=(8, 6))
plt.plot(x_zoom, pdf_class1_zoom, label='Class 1 (variance=1)', color='blue', linewidth=3)
plt.plot(x_zoom, pdf_class2_zoom, label='Class 2 (variance=4)', color='red', linewidth=3)
plt.plot(x_zoom, discriminant_zoom, label='Discriminant Function', color='green', linestyle='--', linewidth=3)

# Adding labels and legend
plt.title('Zoomed-In View of Distributions and Discriminant Function at Intersection')
plt.xlabel('x')
plt.ylabel('Density / Discriminant')
plt.legend()

# Display the zoomed-in plot
plt.grid(True)
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
import numpy as np

# Corrected distance matrix from the image
distance_matrix = np.array([
    [0, 9, 3, 6, 11],  # Point 1
    [9, 0, 7, 5, 10],  # Point 2
    [3, 7, 0, 9, 2],   # Point 3
    [6, 5, 9, 0, 8],   # Point 4
    [11, 10, 2, 8, 0]  # Point 5
])

# Perform hierarchical clustering using single linkage
linked = sch.linkage(sch.distance.squareform(distance_matrix), method='single')

# Plot the dendrogram
plt.figure(figsize=(8, 6))
dendrogram = sch.dendrogram(linked, labels=['Point 1', 'Point 2', 'Point 3', 'Point 4', 'Point 5'])
plt.title('Hierarchical Clustering Dendrogram (Single-Linkage)')
plt.xlabel('Points')
plt.ylabel('Distance')
plt.grid(True)
plt.show()
```
